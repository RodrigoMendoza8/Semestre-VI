```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Rend_IPC (Y)
Y = np.array([0.012, 0.008, -0.004, 0.015, 0.010, 0.006])
# Rend_TC (X1)
X1 = np.array([0.010, 0.015, 0.020, -0.005, 0.000, 0.008])
# Tasa (%) (X2)
X2 = np.array([7.25, 7.50, 7.75, 7.00, 7.25, 7.40])
# Volatilidad (X3)
X3 = np.array([0.020, 0.030, 0.045, 0.018, 0.022, 0.028])
```

```python
x_n = np.column_stack((X1, X2, X3))
model = LinearRegression()
model.fit(x_n, Y)
print("Intercepto:", model.intercept_)
print("Coeficientes:", model.coef_)
mse = np.mean((model.predict(x_n) - Y) ** 2)
print("Error cuadrático medio:", mse)
```

## 2. Coeficientes
El rendimiento de la tasa de cambio es positiva y con un valor de 0.20, lo que significa que si tiene impacto positivo en el rendimiento del IPC.
El porcenataje de la tasa tiene una relacion negativa pero vale 0.01, casi no impacta en el IPC.
La volatilidad de -0.46 impacta de gran manera en lo negativo, a mayor volatilidad menor IPC.

# 3. Valor con mayor impacto
La volatilidad, es el que tiene el valor mas grande aunque impacta de manera negativa.
