```python
#Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#R_IPC
Y = np.array([0.012, 0.008, -0.004, 0.015, 0.010])
# R_TC
X1 = np.array([0.010, 0.015, 0.020, -0.005, 0.000])
# Tasa
X2 = np.array([7.25, 7.50, 7.75, 7.00, 7.25])
# Volatilidad
X3 = np.array([0.020, 0.030, 0.045, 0.018, 0.022])
```

```python
x_n = np.column_stack((X1, X2, X3))
model = LinearRegression()
model.fit(x_n, Y)
print("Intercepto:", model.intercept_)
print("Coeficientes:", model.coef_)
mse = np.mean((model.predict(x_n) - Y) ** 2)
print("Error cuadrático medio:", mse)
```
