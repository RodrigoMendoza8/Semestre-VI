El precio internacional del petróleo impacta los costos de producción y, por tanto, la inflación en
economías importadoras.

* Graficar los datos y describir la relación.
* Estimar el modelo de regresión lineal simple.
* Interpretar el intercepto (inflación base).
* Interpretar el residuo del último año.

```python
import matplotlib.pyplot as plt
import numpy as np

def predecir(x):
    return b0 + x*b1

def error(y, x):
    y_gorro = (b1*x) + b0
    error = y - y_gorro
    return error

precio_petroleo = [45, 55, 65, 75, 85]
inflacion = [3.1, 3.6, 4.0, 4.6, 5.1]

promedio_x = sum(precio_petroleo) / len(precio_petroleo)
promedio_y = sum(inflacion) / len(inflacion)

numerador = 0
denominador = 0

for i in range(len(precio_petroleo)):
    numerador += ((precio_petroleo[i] - promedio_x) * (inflacion[i] - promedio_y)) 
    denominador += (precio_petroleo[i] - promedio_x)**2
    
b1 = numerador / denominador
b0 = promedio_y - b1*promedio_x

print(f"La ecuación del modelo es: y = {b0:.2f} + {b1:.2f}x ")
print(f"Error para el ultimo año (y = {inflacion[4]}) (y gorro = {predecir(precio_petroleo[4]):.2f}): {error(inflacion[4], precio_petroleo[4]):.2f}")
```

## Interpretacion del intercepto
Cuando el precio del petroleo sea 0, la inflacion sera de 0.83%

## Residuo del ultimo año
El error calculado para el ultimo año fue de 0.02, indicando una gran precision a la hora de predecir ese dato.

```python
plt.figure()
plt.scatter(precio_petroleo, inflacion, label='Datos')
x = np.array(precio_petroleo)
y = b1 * x + b0
plt.plot(x, y, color='red', label='Línea de regresión')
plt.xlabel('Precio del Petróleo')
plt.ylabel('Inflación')
plt.legend()
plt.grid()
plt.show()
```

## Grafico
Aunque la pendiente positiva ya nos decia la relacion, al momento de ver el grafico podemos ver que por cada unidad que aumente el precio del petroleo, la inflacion aumentara un 0.05%
