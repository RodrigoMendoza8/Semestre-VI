El Banco de México analiza cómo los cambios en la tasa de interés afectan la inversión privada, clave
para el crecimiento económico.

· Estimar una regresión lineal simple entre tasa e inversión. \
· Interpretar el signo de la pendiente. \
· Explicar el significado económico de un ß1<0. \
· Calcular el consumo estimado para una tasa de 5.8%.

```python
import matplotlib.pyplot as plt
import numpy as np

def predecir(x):
    return b0 + x*b1

tasa_interes = [4.5, 5.0, 5.5, 6.0, 6.5]
inversion_privada = [102, 98, 94, 90, 87]

promedio_x = sum(tasa_interes) / len(tasa_interes)
promedio_y = sum(inversion_privada) / len(inversion_privada)

numerador = 0
denominador = 0

for i in range(len(tasa_interes)):
    numerador += ((tasa_interes[i] - promedio_x) * (inversion_privada[i] - promedio_y)) 
    denominador += (tasa_interes[i] - promedio_x)**2
    
b1 = numerador / denominador
b0 = promedio_y - b1*promedio_x

print(f"La ecuación del modelo es: y = {b0:.2f} + {b1:.2f}x ")
print(f"Para una tasa de interes de 5.8%, la inversion privada va a ser de: {predecir(5.8):.2f}")
```

## Signo de la pendiente
La ecuacion de la recta tiene -7.6 de pendiente, lo que significa que la relacion entre la tasa de interes y la inversion es inversa, va hacia abajo. Entre mayor sea la tasa, menor va a ser la inversion.

## Significado economico de $\beta_1 < 0$
Por cada unidad de la variable independiente, n unidades se le va a restar a la variable dependiente.

```python
plt.figure()
plt.scatter(tasa_interes, inversion_privada, label='Datos')
x = np.array(tasa_interes)
y = b1 * x + b0
plt.plot(x, y, color='red', label='Línea de regresión')
plt.xlabel('Tasa de Interés')
plt.ylabel('Inversión Privada')
plt.legend()
plt.grid()
plt.show()
```
